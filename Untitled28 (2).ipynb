```python
import statistics
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

plddt_scores = []
residue_nums = []
seen = set()

with open("ABL1_alphafold.pdb") as f:
    for line in f:
        if line.startswith('ATOM') and line[12:16].strip() == 'CA':
            res = int(line[22:26].strip())
            if res not in seen:
                plddt_scores.append(float(line[60:66].strip()))
                residue_nums.append(res)
                seen.add(res)

avg   = statistics.mean(plddt_scores)
vh    = sum(1 for s in plddt_scores if s > 90)
hi    = sum(1 for s in plddt_scores if s > 70)
total = len(plddt_scores)

print("=== AlphaFold2 pLDDT Confidence ===")
print("Total residues :", total)
print("Average pLDDT  :", round(avg, 2))
print(f"Very high >90  : {vh} ({100*vh//total}%)")
print(f"High >70       : {hi} ({100*hi//total}%)")

kd = [s for r,s in zip(residue_nums,plddt_scores) if 242<=r<=492]
if kd:
    print("Kinase dom avg :", round(statistics.mean(kd), 2), "(residues 242-492)")

# Plot
colors = ['#0053D6' if s>=90 else '#65CBF3' if s>=70 else '#FFDB13' if s>=50 else '#FF7D45'
          for s in plddt_scores]

fig, ax = plt.subplots(figsize=(14, 4))
ax.bar(residue_nums, plddt_scores, color=colors, width=1.0)
ax.axhline(90, color='#0053D6', ls='--', lw=0.8)
ax.axhline(70, color='#65CBF3', ls='--', lw=0.8)
ax.axhline(50, color='#FFDB13', ls='--', lw=0.8)
ax.axvspan(242, 492, alpha=0.1, color='green')

patches = [
    mpatches.Patch(color='#0053D6', label='>90 Very high'),
    mpatches.Patch(color='#65CBF3', label='70-90 High'),
    mpatches.Patch(color='#FFDB13', label='50-70 Low'),
    mpatches.Patch(color='#FF7D45', label='<50 Very low'),
    mpatches.Patch(color='green', alpha=0.3, label='Kinase domain 242-492')
]
ax.legend(handles=patches, fontsize=8, loc='lower right')
ax.set_xlabel('Residue Number')
ax.set_ylabel('pLDDT Score')
ax.set_title('ABL1 AlphaFold2 pLDDT Confidence (AF-P00519-F1)')
ax.set_ylim(0, 100)
plt.tight_layout()
plt.savefig('ABL1_pLDDT_plot.png', dpi=150)
plt.show()
print("Saved: ABL1_pLDDT_plot.png")
```
